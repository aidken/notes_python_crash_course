```python
import matplotlib.pyplot as plt

squares = (1, 4, 9, 16, 25, 36)

fig, ax = plt.subplots()
ax.plot(squares)

plt.show()
```

```python
fig, ax = plt.subplots()
# fig = entire figure or collectoin of plots
# ax = specific plot
ax.plot(squares, linewidth = 5)
ax.set_title('Square Numbers', fontsize=24)
ax.set_xlabel('Value')
ax.set_ylabel('Square of Value')
ax.tick_params(axis='both', labelsize=14)

plt.show()
```

```python
import math

squares = (1, 4, 9, 16, 25, 36)
input_values = [math.sqrt(i) for i in squares]

fig, ax = plt.subplots()
# fig = entire figure or collectoin of plots
# ax = specific plot
ax.plot(input_values, squares, linewidth=5)
ax.set_title('Square Numbers', fontsize=24)
ax.set_xlabel('Value', fontsize=18)
ax.set_ylabel('Square of Value', fontsize=18)
ax.tick_params(axis='both', labelsize=14)

plt.show()
```
